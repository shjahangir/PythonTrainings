**Covid19 Visualizations Using MatplotLib**

```python
import pandas as pd
from matplotlib import pyplot as plt 
from google.colab import files
uploaded = files.upload()


import io
data = pd.read_csv(io.BytesIO(uploaded['coviddata.csv']))


```

```python
daycount = data['Day']
MY = data['Malaysia']
ID = data['Indonesia']
TH = data['Thailand']
CA = data['Canada']
SG = data['Singapore']
PH = data['Philippines']
JP = data['Japan']
```

```python
plt.rcdefaults()
plt.style.use('fivethirtyeight')
plt.rcParams["axes.formatter.limits"] = (-5, 12)
```

```python
colors = ['#1f77b4', '#aec7e8', '#ff7f0e', '#ffbb78', '#2ca02c', '#98df8a']


plt.plot(daycount,MY,color=colors[0],label='Malaysia')
plt.plot(daycount,ID,color=colors[1],label='Indonesia')
plt.plot(daycount,TH,color=colors[2],label='Thailand')
plt.plot(daycount,SG,color=colors[3],label='Singapore')
plt.plot(daycount,PH,color=colors[4],label='Philippines')
plt.plot(daycount,JP,color=colors[5],label='Japan')


plt.legend()
plt.title('Covid Infection in Asia')
plt.ylabel('Number of Infections')
plt.xlabel('Number of Days')

plt.savefig("plot.png")

plt.show()
```
